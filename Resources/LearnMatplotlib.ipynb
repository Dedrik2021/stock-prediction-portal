# Type of plots

## Line plots

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
plt.plot(x, y);
plt.title('Simple Line Plot');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True);
```

## Chat Bar

```python
mounts = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(mounts, revenue);
plt.title('Mountly Revenue Bar Chat');
plt.xlabel('Mounts');
plt.ylabel('Revenue');
```

## Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 25, 20, 30, 40]
predicted_price = [8, 21, 25, 30, 38]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.title('Plot With Legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

## Subplots

```python
plt.figure(figsize=(10, 6))
# First subplot - original price
plt.subplot(3, 1, 1);
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.title('Original price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();


# Second subplot - Predicted Price
plt.subplot(3, 1, 2);
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('Predicted price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

# Third subplot - Difference between Original & Predicted Prices
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3);
plt.bar(days, price_difference, color='purple', label='Price Difference');
plt.title('Difference between Original & Predicted Prices');
plt.xlabel('Days');
plt.ylabel('Price Difference');
plt.legend();

plt.tight_layout()
```

## Saving the figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.title('Plot With Legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.savefig('example_prediction.png', dpi=300) # save the image 
```
